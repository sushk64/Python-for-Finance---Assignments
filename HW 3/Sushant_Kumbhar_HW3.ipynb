```python
#Question 1

class MovingAverage:

    def __init__(self, size):
        self.size = size  #Initializing the variable size
        self.Moving_Average = []  #Initializing the variable ls as empty list
    
    def next(self, value):
        self.Moving_Average.append(value)  #Appending the values in the empty list
        if len(self.Moving_Average) <= self.size:  #Calculating moving average by giving condition to check if list size less than or equal to window size
            x = round(sum(self.Moving_Average) / len(self.Moving_Average),5)
            return x 
        else:
            Moving_Average = self.Moving_Average[-self.size:] 
            y = round(sum(Moving_Average) / len(Moving_Average),5)
            return y
```

```python
#Question 2

class subway:

    def __init__(self):
        self.check_in = {}  #Initialize the variable check_in
        self.check_out = {}  #Initialize the variable check_out
        
        
    def checkIn(self, id, stationName, t):     
        if stationName in self.check_in:  #Condition to store check_in station name and time into the dictionary
            self.check_in[stationName].append([id, t]) 
        else:
            self.check_in[stationName] = [[id, t]]      
    
    def checkOut(self, id, stationName, t):
        if stationName in self.check_out:  #Condition to store check_out station name and time into the dictionary
            self.check_out[stationName].append([id, t])
        else:
            self.check_out[stationName] = [[id, t]]
    
    def getAverageTime(self, startStation, endStation):
        average_time = []  #Creating the empty list to store the time of check_in and check_out
        start_station = self.check_in[startStation]
        end_station = self.check_out[endStation]
        for i,j in zip(start_station,end_station):  #Calculating the average time taken between start and end
                if i[0] == j[0]:
                    average_time.append(j[1] - i[1])
                    z = round(float(sum(average_time) / len(average_time)),3)
        return z
```

```python
#Question 3

class Linear_regression:

    def __init__(self, x, y, m, c, epochs, L): 
       self.x,self.y,self.m,self.c,self.epochs,self.L = x,y,m,c,epochs,L #Initializing variables x,y,m,c,epochs,L
              
    def gradient_descent(self) : 
        for it in range(0,self.epochs):
            Dm = []  # create an empty list
            Dc = []  # create an empty list
            for i in range(len(self.x)):  # creating for loop to iterate through length of list x
                y_pred = (x[i][0] * self.m) + self.c
                Dm.append((x[i][0] * (y_pred - y[i])))  # append new values of Dm for every iteration
                Dc.append((y_pred - y[i]))            
            dm = sum(Dm) / len(Dm)  
            dc = sum(Dc) / len(Dc)  
            self.m = self.m - (self.L * dm)
            self.c = self.c - (self.L * dc)
        return self.m, self.c
    
    
    def predict(self,new_x):  #Creating function to predict values of m and c  
       predicted_ls = []
       for j in new_x:
           predicted_ls.append((self.m * j) + self.c)
       return predicted_ls
```
